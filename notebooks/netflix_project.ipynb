## Middle Course Project

We will imagine you are working for Netflix, and you are tasked with writting a program that describes how many viewers are currently watching which shows. 

For simplicity, we will imagine that there are only three totally non-biased shows currently available: The Witcher, The Umbrella Academy, One Piece and Brooklyn Nine-Nine.

Unfortunately, we don’t keep an active registry of Active Users by show..

What we do have is a registry of when a User **Starts** watching, and when they **Stop** watching. something like this:

```python

    {
        "show": "The Witcher",
        "event": "start",

    },

        "user_id": 2
    },
    {
        "show": "The Umbrella Academy",
        "event": "start",
        "timestamp": 4,
        "user_id": 1
    },
    {
        "show": "The Umbrella Academy",
        "event": "stop",
        "timestamp": 5,
        "user_id": 1
    },
]
```

*Notes*  on the current data:

- Timestamps are unique.
- User_ids are unique.
- Two types of events: “start” and “stop” only.

### Project Objective:

Write a program that outputs a report of the number of currently active users for each show. 

### Program Input:

The program should receive as only input a registry list of user events, such that each event is a dictionary with four entries: “show”, “event”, “timestamp” and “user_id”.

### Program Output:

The output of running the program should be: 

- wrapped in a function that takes the event list as argument.
- one line by Netflix show only, formatted to have the **name of the show** and the **number of users**.
- written to **stdout**.

**Note** Don’t start by writing “function”. make it work, then wrap it.

**Note** Develop using the provided data, but the program should work with any data.
